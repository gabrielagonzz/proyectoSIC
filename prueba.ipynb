```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('cripto_monedas.csv')

df.index = df.index.astype(str).str.upper()


df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d', errors='coerce')

df['date'] = pd.to_datetime(df['date'])

# Eliminar filas con valores nulos
df = df.dropna(subset=['date'])
df = df.dropna(subset=['market_cap'])

# Identificar valores faltantes o nulos
valores_nulos = df.isnull().sum()
print("Valores nulos por columna:\n", valores_nulos)

df['total_volume'] = pd.to_numeric(df['total_volume'], errors='coerce')
df['market_cap'] = pd.to_numeric(df['market_cap'], errors='coerce')

# Convertir los índices a mayúsculas
df.index = df.index.str.upper()

# Separar por grupos dependiendo del valor de mercado en la fecha 1 de Enero de 2015
df['year'] = df['date'].dt.year
df_enero_2015 = df[df['date'] == '2015-01-01']

# Organizar por valor ascendente
df_enero_2015 = df_enero_2015.sort_values(by='market_cap', ascending=True)

grupos_por_valor_mercado = df_enero_2015.groupby('market_cap')

for grupo, datos in grupos_por_valor_mercado:
    print(datos)

```

```python
# Eliminar filas con valores nulos en la columna 'date'
df = df.dropna(subset=['date'])

# Verificar si hay valores nulos después de eliminarlos
valores_nulos_despues = df[['date', 'price']].isnull().sum()

import pandas as pd
import matplotlib.pyplot as plt

# Filtrar datos del año 2015 y las 4 criptomonedas más interesantes
df_2015 = df[df['date'].dt.year == 2015]
criptomonedas_interesantes = ['bitcoin', 'ethereum', 'ripple', 'litecoin']
df_2015_top4 = df_2015[df_2015['coin_name'].isin(criptomonedas_interesantes)]

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Precio de las 4 criptomonedas más interesantes en 2015', fontsize=16)

for i, criptomoneda in enumerate(criptomonedas_interesantes):
    fila = i // 2
    columna = i % 2
    datos_criptomoneda = df_2015_top4[df_2015_top4['coin_name'] == criptomoneda]
    axs[fila, columna].plot(datos_criptomoneda['date'], datos_criptomoneda['price'])
    axs[fila, columna].set_title(criptomoneda)
    axs[fila, columna].set_xlabel('Fecha')
    axs[fila, columna].set_ylabel('Precio')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('cripto_monedas.csv')

df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d', errors='coerce')

df_2015 = df[df['date'].dt.year == 2015]

# Media de las criptomonedas y la desviación estándar menor entre ellas
media = df_2015.groupby('coin_name')['price'].mean()
dev_est = df_2015.groupby('coin_name')['price'].std()

dev_est_baja = dev_est.idxmin()
print(f"La criptomoneda con la desviación estándar más baja en 2015 es {dev_est_baja}.\n")

# Las criptomonedas que están por encima de la media
toda_media = df_2015['price'].mean()
above_media = media[media > toda_media]
print(f"La criptomoneda que están por encima de la media en 2015 es {above_media}\n")

# Criptomonedas que tuvieron un mayor valor en el mercado el los ult. 3 meses
last_three_months = df_2015[df_2015['date'].dt.month >= 10]
max_last_three_months = last_three_months.groupby('coin_name')['price'].max()
max_value_cryptos = max_last_three_months.sort_values(ascending=False).head()
print("Las criptomonedas con mayor valor en los últimos tres meses del año 2015 fueron:")
print(max_value_cryptos)

# La criptomoneda mas volatil
volatilidad = df_2015.groupby('coin_name')['price'].std().idxmax()
print("La criptomoneda más volátil en 2015 fue:", volatilidad)
```
